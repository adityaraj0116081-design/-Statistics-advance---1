1.  What is a random variable in probability theory?

> A random variable is a function that assigns a numerical value to each possible outcome of a random experiment. It allows us to describe uncertain events using numbers, making them easier to analyze mathematically.

Definition

A random variable (X) is a variable whose value depends on the outcome of a random experiment.

Example 1: Tossing a Coin

Suppose you toss a coin once.

Head (H) → X=1
Tail (T) → X=0

Here, X is the random variable because it assigns a number to each outcome.


2. What are the types of random variables?

> There are two main types of random variables in probability theory:

1. Discrete Random Variable

A discrete random variable can take only a countable number of distinct values.

Characteristics:

Takes separate, countable values (0, 1, 2, 3, ...)
Used for counting outcomes
Has a Probability Mass Function (PMF)

Examples:

Number of heads in 3 coin tosses (0, 1, 2, or 3)
Number of students in a class
Number of defective items in a batch
2. Continuous Random Variable

A continuous random variable can take any value within a given range or interval.

Characteristics:

Takes infinitely many possible values
Used for measuring quantities
Has a Probability Density Function (PDF)

Examples:

Height of a person
Weight of an object
Temperature
Time taken to complete a task


3. Explain the difference between discrete and continuous distributions.


> The main difference between discrete and continuous probability distributions is the type of values the random variable can take.

6.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?

> The Central Limit Theorem (CLT) is one of the most important concepts in statistics. It states that as the sample size becomes large, the distribution of the sample means approaches a normal (bell-shaped) distribution, regardless of the shape of the original population distribution, provided the samples are random and independent.

Suppose you repeatedly take random samples of size n from a population and calculate the mean of each sample.

If the sample size is sufficiently large (commonly n ≥ 30), the distribution of these sample means will be approximately normal.
This holds even if the original population is not normally distributed.
Key Conditions
Samples are selected randomly.
Observations are independent.
The sample size is sufficiently large (typically 30 or more).

7. What is the significance of confidence intervals in statistical analysis?

> A confidence interval (CI) is a range of values that is likely to contain the true value of a population parameter (such as the population mean or proportion) based on sample data.

Significance of Confidence Intervals
Estimates Population Parameters
Instead of giving a single estimate, a confidence interval provides a range in which the true population value is likely to lie.
Indicates Precision
A narrow confidence interval indicates a more precise estimate.
A wide confidence interval indicates greater uncertainty.
Measures Reliability
The confidence level (e.g., 95%) indicates how confident we are in the estimation method.
A 95% confidence interval means that if the same sampling process were repeated many times, about 95% of the calculated intervals would contain the true population parameter.
Supports Decision-Making
Confidence intervals help researchers and businesses assess whether results are meaningful and make informed decisions.
Used in Hypothesis Testing
Confidence intervals complement hypothesis tests.
For example, if a 95% confidence interval for a difference in means does not include 0, it suggests a statistically significant difference at the 5% significance level.


8. What is the concept of expected value in a probability distribution?


> The expected value (also called the mean or expectation) of a probability distribution is the long-run average value of a random variable if the same experiment is repeated many times.

It represents the average outcome you would expect over a large number of trials.

Formula for a Discrete Random Variable

The expected value is calculated as:

E(X)=∑x⋅P(X=x)

where:

x = possible values of the random variable
P(X=x) = probability of each value


```python
# # 9. Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Compute mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

# Display the results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)

# Draw histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
